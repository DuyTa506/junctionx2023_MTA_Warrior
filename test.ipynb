```python
import math
import random
import matplotlib.pyplot as plt

def gift_wrapping(points):
    """Computes the convex hull of a set of 2D points using the Gift wrapping algorithm."""
    
    # Find the point with the lowest y-coordinate (and lowest x-coordinate if there are ties)
    start = min(points, key=lambda p: (p[1], p[0]))
    
    # Initialize the hull with the starting point
    hull = [start]
    
    # Loop until we return to the starting point
    while True:
        # Find the point with the smallest polar angle from the last point on the hull
        last_point = hull[-1]
        next_point = None
        for point in points:
            if point == last_point:
                continue
            if next_point is None:
                next_point = point
            angle1 = math.atan2(point[1] - last_point[1], point[0] - last_point[0])
            angle2 = math.atan2(next_point[1] - last_point[1], next_point[0] - last_point[0])
            if angle1 < angle2:
                next_point = point
                
        # If the next point is the starting point, we're done
        if next_point == start:
            break
        
        # Add the next point to the hull
        hull.append(next_point)
    
    # Add the starting point to the hull to complete the loop
    hull.append(start)
    
    # Return the list of points on the convex hull
    return hull

# Generate some random 2D points
points = [(random.uniform(-1, 1), random.uniform(-1, 1)) for i in range(50)]

# Compute the convex hull
hull = gift_wrapping(points)

# Plot the points and the hull
plt.scatter([p[0] for p in points], [p[1] for p in points])
plt.plot([p[0] for p in hull], [p[1] for p in hull], 'r')
plt.show()
```

```python
import cv2
import numpy as np

# Create a set of random 2D points
points = np.random.randint(0, 100, size=(10,2))

# Compute the convex hull
hull = cv2.convexHull(points)

# Draw the convex hull on an image
img = np.zeros((100, 100, 3), dtype=np.uint8)
cv2.drawContours(img, [hull], 0, (0, 255, 0), thickness=2)

# Display the image
cv2.imshow('Convex Hull', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
